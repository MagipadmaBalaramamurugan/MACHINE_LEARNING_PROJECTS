```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'C:\Users\MAGI\OneDrive\Documents\DATASET\diabetes2.csv') 
```

```python
df.head()
```

```python
print("Shape:", df.shape)
```

```python
label_encoder = LabelEncoder()
for col in df.columns:
    df[col] = label_encoder.fit_transform(df[col])

df.head()
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

```python
X = df.drop(columns=['Outcome'], axis=1)
Y = df['Outcome']
print(X)
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

print("Standardized features preview:")
print(X[:5])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, stratify=Y, random_state=2)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, Y_train)
```

```python
from sklearn import svm

classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)

print("Model training completed.")
```

```python
from sklearn.metrics import accuracy_score
train_preds = classifier.predict(X_train)
train_acc = accuracy_score(train_preds, Y_train)
print("Training accuracy:", train_acc)
test_preds = classifier.predict(X_test)
test_acc = accuracy_score(test_preds, Y_test)
print("Test accuracy:", test_acc)
```

```python
import numpy as np
import pandas as pd
input_data = (10,188,64,0,26.3,0.652,72)
input_df = pd.DataFrame([input_data], columns=[
    'Pregnancies', 'Glucose', 'BloodPressure',
    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'
])
input_std = scaler.transform(input_df)
prediction = classifier.predict(input_std)
print("PREDICTION:", prediction)
print("RESULT: THE PATIENT IS ", "DIABETIC" if prediction[0] == 1 else "NOT DIABETIC")
```
